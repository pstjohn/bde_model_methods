```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='talk', style='ticks', color_codes=True)
from tqdm._tqdm_notebook import tqdm_notebook
tqdm_notebook.pandas()

%matplotlib inline
```

```python
runs = ['layer_runs/n_message_{}'.format(i) for i in range(1, 11)]

fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(8,4))

ax[0].set_yscale('log')
ax[1].set_yscale('log')

logs = {}

for run in runs:
    
    try:
        logs[run] = log = pd.read_csv('{}/log.csv'.format(run), index_col=0)
    except pd.EmptyDataError:
        pass
    
    ax[0].plot(log.loss)
    ax[1].plot(log.val_loss)

# ax[1].legend(loc='center left', bbox_to_anchor=(1., .5))
sns.despine()

ax[0].set_xlabel('Epoch')
ax[1].set_xlabel('Epoch')

ax[0].set_ylabel('Loss (MAE)')

ax[0].set_title('Training')
ax[1].set_title('Validation')

min_loss = min([log['val_loss'].min() for log in logs.values()])
ax[1].axhline(min_loss, color='.6', ls='--')
ax[1].text(0, min_loss, '\nMAE={:.3f}'.format(min_loss), va='top')

plt.ylim([0.1, 10])
```
